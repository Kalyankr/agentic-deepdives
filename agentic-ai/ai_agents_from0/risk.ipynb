```python
UniqueEventTypes_Attr = [
    "PostAuth",
    "UpdateRiskEvaluation",
    "SignUpFeedback",
    'AddPIFeedback',
]

EventType_List_EventType = [
    "",
    "behavior_analysis",
    "escalation_analysis",
    "purchase",
    "signup",
    "appendtoken",
    "addpi",
    "threshold_payment",
    "updatepi",
    "deviceexchange",
    "redeem",
    "create_subscription",
    "policy",
    "payout",
    "recurring_purchase"
]
 

EventType_List_FeedType = [
    "AddAuthResult",
    "AddPIResult",
    "AddManualReviewResult"
]


All_EventTypes =   UniqueEventTypes_Attr + EventType_List_EventType + EventType_List_FeedType
len(All_EventTypes)
```

```python
# Convert to lowercase for case-insensitive comparison
all_eventypes_lower = [event.lower() for event in All_EventTypes]
eventtype_list_pii_lower = [event.lower() for event in EventType_List_Pii]

# Find existing values
existing = [event for event in All_EventTypes if event.lower() in eventtype_list_pii_lower]

# Find missing values
missing = [event for event in All_EventTypes if event.lower() not in eventtype_list_pii_lower]

print("Existing in EventType_List_Pii:")
print(existing)
print(f"\nCount: {len(existing)}")

print("\n" + "="*50)
print("\nMissing from EventType_List_Pii:")
print(missing)
print(f"\nCount: {len(missing)}")
```

```python
# What's in EventType_List_Pii but MISSING from All_EventTypes
all_eventypes_lower = [event.lower() for event in All_EventTypes]

missing_from_all = [event for event in EventType_List_Pii if event.lower() not in all_eventypes_lower]

print("Missing from All_EventTypes (present in EventType_List_Pii but not in All_EventTypes):")
for i, event in enumerate(missing_from_all, 1):
    print(f"  {i}. {event}")
print(f"\nTotal missing: {len(missing_from_all)}")
```
